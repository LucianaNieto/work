```python
def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days)):
        yield start_date + timedelta(n)
```

### 2. Automated creation of manifiest.json 

```python
#project_name = "projects/earthengine-legacy/assets/users/luzluz2054/2017"
#project_name = "projects/phenology-052020/assets/cambodia_2020_12E"
#project_name = "projects/phenology-052020/assets/cambodia_2021_12E"
#project_name = "projects/phenology-052020/assets/cambodia_2020_13E"
#project_name = "projects/phenology-052020/assets/cambodia_2021_13E"
for single_date in daterange(date(2020, 1, 1), date(2021, 1, 1)):   #change this to extract only 2020 or only 2021
    dayintheyear = single_date.strftime("%Y-%m-%d")
    year = single_date.year
    month = single_date.strftime("%m")
    day = single_date.strftime("%d")
    
    #day = single_date.day
    start_time = int(datetime(single_date.year, single_date.month, single_date.day, 0, 0).timestamp())
    end_time = start_time + 86400
    json_content =  template_manifiest_14N(project_name,dayintheyear,year,month,day,start_time,end_time)
    
    with open('./data-manifests/13E60N_2020/{0}.json'.format(dayintheyear), 'w') as f:  #change the folder to store the json
        f.write(json_content)
```
